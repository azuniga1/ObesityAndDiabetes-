```python
# Dependencies
import pandas as pd
from glob import glob
```

```python
# use glob() to list all files that match a pattern 
obesity_files = sorted(glob('Resources/*-Obesity.csv'))
obesity_files
```

```python
# Concat all files  by using pd.concat() and asssign() methods
obesity_2011_to_2016 = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in obesity_files), ignore_index = True)
obesity_2011_to_2016.head()
```

```python
# Show specific columns used
New_Df = obesity_2011_to_2016[['ID', 'YearStart', 'LocationAbbr', 'LocationDesc', 'Question', 'Data_Value', 'GeoLocation_Lat', 
                'GeoLocation_Long',  ]]
New_Df.head()
```

```python
New_Df.dtypes
```

```python
# Export file as a CSV, without the pandas index, but with the header
New_Df.to_csv('Output/Obesity2011_2016.csv', index=False, header=True)
```
